```python
def convert_string_in(dat):
    columns=dat.columns.values.tolist()
    #print (columns)
    for col in columns:
        ind_list1=list(set(dat[col].values.tolist()))
        #ind_list1=dict(zip(ind_list1, range(len(ind_list1))))
        def to_in(abc):
            return ind_list1.index(abc)
        dat[col]=dat[col].map(to_in)
    return dat
```

```python
df[['INDICATION', 'PAYER_TYPE', 'PHRM_TYPE']]=convert_string_in(df[['INDICATION', 'PAYER_TYPE', 'PHRM_TYPE']])
```

```python
def cast_everything(dat):
    columns=dat.columns.values.tolist()
    for col in columns:
        df[col]=df[col].astype(np.float64)
    return dat
```

```python
def fill_everything(dat):
    columns=dat.columns.values.tolist()
    for col in columns:
        df[col].fillna(df[col].median(), inplace=True)
    return dat
```

```python
df=cast_everything(df)
df=fill_everything(df)
```

```python
def print_col_size(dat):
    columns=dat.columns.values.tolist()
    for col in columns:
        ind1=set(dat[col].values.tolist())
        print (col, len(ind1))
    return dat
```

```python
print_col_size(df)
```

```python
def make_dummies(dat):
    columns=dat.columns.values.tolist()
    for col in columns:
        n=len(set(dat[col].values.tolist()))
        if n<=10:
            tickets_dummies = pd.get_dummies(dat[col],prefix=col)
            dat = pd.concat([dat, tickets_dummies],axis=1)
            dat.drop(col,inplace=True,axis=1)
    return dat
```

```python
def convert_target_int(dataf):
    def target_int(val):
        if val=='P':
            return 0
        if val=='DS':
            return 1
        return 2
    dataf=dataf.map(target_int)
    return dataf
```

```python
def convert_int_target(dataf):
    def target_int(val):
        if val==0:
            return 'P'
        if val==1:
            return 'DS'
        return 'DO'
    dataf['PID_State']=dataf['PID_State'].map(target_int)
    return dataf
```

```python
scaler = preprocessing.StandardScaler().fit(df)
df_try=pd.DataFrame(scaler.transform(df))
```

```python
X_train=df_try.ix[0:5000-1]
X_test=df_try.ix[5000:]
print (X_train.shape)
print (X_test.shape)
```

```python
from sklearn.svm import LinearSVC
from sklearn.ensemble import ExtraTreesClassifier
clf = ExtraTreesClassifier(n_estimators=200)
#clf = LinearSVC(C=0.01, penalty="l1", dual=False)
clf = clf.fit(X_train, target2)
features = pd.DataFrame()
features['feature'] = X_train.columns
features['importance'] = clf.feature_importances_
important_features=features.sort('importance', ascending=False)['feature'].values.tolist()
important_features=important_features[0:25]
print(important_features)
print(features)
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.feature_selection import SelectKBest
model = SelectFromModel(clf, prefit=True)
train_new = model.transform(X_train)
print(train_new.shape)
test_new=model.transform(X_test)
print (test_new.shape)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import StratifiedKFold
from sklearn.grid_search import GridSearchCV
from sklearn.ensemble.gradient_boosting import GradientBoostingClassifier
from sklearn.cross_validation import cross_val_score
forest = RandomForestClassifier(max_features='sqrt')
gb=GradientBoostingClassifier(max_features='sqrt')
dt=DecisionTreeClassifier(max_features='sqrt')
parameter_grid_randomForest = {
                 'max_depth' : [4,5,6,7,8],
                 'n_estimators': [200,210,240,250],
                 'criterion': ['gini','entropy']
                 }

cross_validation = StratifiedKFold(np.array(target2), n_folds=5)

grid_search = GridSearchCV(forest,
                           param_grid=parameter_grid_randomForest,
                           cv=cross_validation)

grid_search.fit(train_new, np.array(target2))

print('Best score: {}'.format(grid_search.best_score_))
print('Best parameters: {}'.format(grid_search.best_params_))
```
